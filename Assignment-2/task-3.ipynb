```python
import math
def calculate_area(shape, **kwargs):
        """
        Calculates the area of various shapes (circle, rectangle, triangle).

        Args:
            shape (str): The type of shape ('circle', 'rectangle', 'triangle').
            **kwargs: Keyword arguments containing the necessary dimensions.

        Returns:
            float or str: The calculated area or an error message.
        """
        shape = shape.lower()  # Standardize input case

        if shape == 'circle':
            # Area of a circle: A = pi * r^2
            if 'radius' in kwargs:
                radius = kwargs['radius']
                return math.pi * (radius ** 2)
            else:
                return "Error: Circle requires 'radius'."

        elif shape == 'rectangle':
            # Area of a rectangle: A = length * width
            if 'length' in kwargs and 'width' in kwargs:
                length = kwargs['length']
                width = kwargs['width']
                return length * width
            else:
                return "Error: Rectangle requires 'length' and 'width'."

        elif shape == 'triangle':
            # Area of a triangle: A = 0.5 * base * height
            if 'base' in kwargs and 'height' in kwargs:
                base = kwargs['base']
                height = kwargs['height']
                return 0.5 * base * height
            else:
                return "Error: Triangle requires 'base' and 'height'."

        else:
                    return f"Error: Unknown shape type '{shape}'. Supported types are 'circle', 'rectangle', 'triangle'."
        
    
        # Example usage:
        print(calculate_area('circle', radius=5))
        print(calculate_area('rectangle', length=10, width=4))
        print(calculate_area('triangle', base=6, height=8))

```
