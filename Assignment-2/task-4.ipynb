```python
def sum_of_squares(numbers: list) -> int or float:
    """
    Calculates the sum of the squares of all numbers in a given list.

    Args:
        numbers (list): A list of numerical values (integers or floats).

    Returns:
        int or float: The calculated sum of the squares.
    """
    total_sum = 0
    # Iterate through each number in the input list
    for num in numbers:
        # Square the number and add it to the running total
        total_sum += num ** 2
    
    return total_sum

# --- Example Usage ---

# 1. Test Case with Integers
list_a = [1, 2, 3, 4]
# Expected: 1^2 + 2^2 + 3^2 + 4^2 = 1 + 4 + 9 + 16 = 30
result_a = sum_of_squares(list_a)
print(f"List: {list_a} | Sum of Squares: {result_a}") 

# 2. Test Case with Floats and Negative Numbers
list_b = [-2.0, 1.5, 3]
# Expected: (-2)^2 + (1.5)^2 + (3)^2 = 4 + 2.25 + 9 = 15.25
result_b = sum_of_squares(list_b)
print(f"List: {list_b} | Sum of Squares: {result_b}")

def sum_of_squares(numbers: list) -> int or float:
    """
    Calculates the sum of the squares of all numbers in a given list.

    Args:
        numbers (list): A list of numerical values (integers or floats).

    Returns:
        int or float: The calculated sum of the squares.
    """
    total_sum = 0
    # Iterate through each number in the input list
    for num in numbers:
        # Square the number and add it to the running total
        total_sum += num ** 2
    
    return total_sum

# --- Example Usage ---

# 1. Test Case with Integers
list_a = [1, 2, 3, 4]
# Expected: 1^2 + 2^2 + 3^2 + 4^2 = 1 + 4 + 9 + 16 = 30
result_a = sum_of_squares(list_a)
print(f"List: {list_a} | Sum of Squares: {result_a}") 

# 2. Test Case with Floats and Negative Numbers
list_b = [-2.0, 1.5, 3]
# Expected: (-2)^2 + (1.5)^2 + (3)^2 = 4 + 2.25 + 9 = 15.25
result_b = sum_of_squares(list_b)
print(f"List: {list_b} | Sum of Squares: {result_b}")
```
