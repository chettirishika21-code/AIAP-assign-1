```python
def calculate_fibonacci(n: int) -> int:
    """
    Calculate the n-th Fibonacci number using a simple recursive approach.
    Zero-based indexing: F(0) = 0, F(1) = 1.

    Args:
        n (int): Index (n >= 0) of the Fibonacci sequence to compute.

    Returns:
        int: The n-th Fibonacci number.

    Raises:
        ValueError: If n is negative.
    """
    if n < 0:
        raise ValueError("n must be a non-negative integer")

    # Base cases: F(0) = 0, F(1) = 1
    if n == 0:
        return 0
    if n == 1:
        return 1

    # Recursive case:
    # F(n) = F(n-1) + F(n-2)
    # The function calls itself twice with smaller arguments until it reaches the base cases.
    return calculate_fibonacci(n - 1) + calculate_fibonacci(n - 2)
```

```python
def fibonacci_iterative(n: int) -> int:
    if n < 0:
        raise ValueError("n must be non-negative")
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a
```
