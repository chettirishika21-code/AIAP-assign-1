```python
def print_multiples_for(n):
    """
    Prints the first 10 multiples of the given number using a for loop.
    """
    for i in range(1, 11):
        print(f"{n} x {i} = {n * i}")

# Example usage:
print_multiples_for(5)
```

```python
def print_multiples_while(n):
    """
    Prints the first 10 multiples of the given number using a while loop.
    """
    i = 1
    while i <= 10:
        print(f"{n} x {i} = {n * i}")
        i += 1

# Example usage:
print_multiples_while(5)
```
