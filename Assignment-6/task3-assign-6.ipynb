```python
def classify_age_group(age):
    """
    Classifies the input age into groups using nested if-elif-else statements.
    Returns a string label for the age group.
    """
    if age < 0:
        return "Invalid age"
    elif age < 13:
        return "Child"
    elif age < 20:
        return "Teenager"
    elif age < 60:
        return "Adult"
    else:
        return "Senior"

# Example usage:
print(classify_age_group(8))    # Child
print(classify_age_group(16))   # Teenager
print(classify_age_group(35))   # Adult
print(classify_age_group(70))   # Senior
```

```python
def classify_age_group_match(age):
    """
    Classifies age using match-case (Python 3.10+).
    """
    match age:
        case a if a < 0:
            return "Invalid age"
        case a if a < 13:
            return "Child"
        case a if a < 20:
            return "Teenager"
        case a if a < 60:
            return "Adult"
        case _:
            return "Senior"

# Example usage:
print(classify_age_group_match(8))    # Child
print(classify_age_group_match(16))   # Teenager
print(classify_age_group_match(35))   # Adult
print(classify_age_group_match(70))   # Senior
```
