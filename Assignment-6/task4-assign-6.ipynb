```python
def sum_to_n_for(n):
    """
    Calculates the sum of the first n natural numbers using a for loop.
    Returns the sum as an integer.
    """
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

# Example usage:
print(sum_to_n_for(10))  # Output: 55
```

```python
def sum_to_n_while(n):
    """
    Calculates the sum of the first n natural numbers using a while loop.
    Returns the sum as an integer.
    """
    total = 0
    i = 1
    while i <= n:
        total += i
        i += 1
    return total

# Example usage:
print(sum_to_n_while(10))  # Output: 55
```

```python
def sum_to_n_builtin(n):
    """
    Calculates the sum of the first n natural numbers using Python's built-in sum and range.
    """
    return sum(range(1, n + 1))

# Example usage:
print(sum_to_n_builtin(10))  # Output: 55
```
